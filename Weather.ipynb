```python
import pandas as pd
import altair as alt

alt.data_transformers.enable("vegafusion")
```

```python
df = pd.read_csv("./weather_classification_data.csv")
df
```

```python
pie_base = (
    alt.Chart(df)
        .transform_aggregate(
            AvgPrecipitation="average(Precipitation (%))",
            groupby=["Location"]
        )
        .encode(
            alt.Theta("AvgPrecipitation:Q").stack(True),
            alt.Color("Location:N")
        )
)

pie_arc = (
    pie_base
        .mark_arc(
            innerRadius=30,
            outerRadius=80
        )
)

pie_labels = (
    pie_base
        .mark_text(radius=105)
        .encode(
            alt.Text("Location:N")
        )
)

pie_percentages = (
    pie_base
        .mark_text(
            radius=52.5
        )
        .transform_calculate(
            AvgPrecipitationHundreth="datum.AvgPrecipitation/100"
        )
        .encode(
            alt.Text("AvgPrecipitationHundreth:Q", format=".2p"),
            color=alt.value("white")
        )
)

pie_arc + pie_labels + pie_percentages
```
